```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters for the three Gaussian distributions
means = [0, 5, 10]  # Means for the distributions
stds = [1, 2, 3]    # Standard deviations for the distributions
num_samples = 1000  # Number of samples for each distribution

# Generate the data for each distribution
data1 = np.random.normal(means[0], stds[0], num_samples)
data2 = np.random.normal(means[1], stds[1], num_samples)
data3 = np.random.normal(means[2], stds[2], num_samples)

# Create a range of x values for plotting the PDFs
x_values = np.linspace(-10, 20, 1000)

# Calculate the Gaussian PDFs using the manual formula
def gaussian_pdf(x, mean, std):
    return (1 / (std * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std) ** 2)

pdf1 = gaussian_pdf(x_values, means[0], stds[0])
pdf2 = gaussian_pdf(x_values, means[1], stds[1])
pdf3 = gaussian_pdf(x_values, means[2], stds[2])

# Plot the distributions
plt.figure(figsize=(10, 6))

# Plot the histograms
plt.hist(data1, bins=50, alpha=0.6, label=f'Mean = {means[0]}, Std = {stds[0]}', density=True)
plt.hist(data2, bins=50, alpha=0.6, label=f'Mean = {means[1]}, Std = {stds[1]}', density=True)
plt.hist(data3, bins=50, alpha=0.6, label=f'Mean = {means[2]}, Std = {stds[2]}', density=True)

# Plot the Gaussian PDFs on top of the histograms
plt.plot(x_values, pdf1, 'k-', lw=2, label=f'Gaussian PDF (Mean = {means[0]}, Std = {stds[0]})')
plt.plot(x_values, pdf2, 'r-', lw=2, label=f'Gaussian PDF (Mean = {means[1]}, Std = {stds[1]})')
plt.plot(x_values, pdf3, 'b-', lw=2, label=f'Gaussian PDF (Mean = {means[2]}, Std = {stds[2]})')

plt.title('Gaussian Distributions with Histograms and PDFs')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

# Counting number of bins 
The number of bins \( k \) is calculated using Sturges' rule as:

$$k = \lceil \log_2 n + 1 \rceil$$

where \( n \) is the number of data points.

Freedman-Diaconis Rule
The number of bins \( k \) is calculated using the Freedman-Diaconis rule as:

$$k = \left\lceil \frac{\text{range}}{2 \times \text{IQR} \times n^{-1/3}} \right\rceil$$

where \( IQR \) is the interquartile range (i.e., the difference between the 75th and 25th percentiles), and \( n \) is the number of data points.

Scott’s Rule
The number of bins \( k \) is calculated using Scott’s rule as:

$$k = \left\lceil \frac{\text{range}}{3.49 \times \text{std} \times n^{-1/3}} \right\rceil$$

where \( std \) is the standard deviation of the data, and \( n \) is the number of data points.

Square-Root Rule
The number of bins \( k \) is calculated using the square-root rule as:

$$k = \lceil \sqrt{n} \rceil$$


## Putting it all Together
* How a combination of Gaussian and bayes theorem makes a powerfull tool like KF is hidden in this chunk of the code. 
* Simply to put, instead of analyzing blue bars (i.e., discrete bayes filter) analyze its counterpart (i,e. a Gussian contineous bell curve in red color* )

```python
def normalize(p):
    return p / sum(p)

def update(likelihood, prior):
    return normalize(likelihood * prior)

prior =      normalize(np.array([4, 2, 0, 7, 2, 12, 35, 20, 3, 2]))
likelihood = normalize(np.array([3, 4, 1, 4, 2, 38, 20, 18, 1, 16]))
posterior = update(likelihood, prior)
#book_plots.bar_plot(posterior)
xs = np.arange(0, 10, .01)
xvals = np.arange(0,10,1)
def mean_var(p):
    x = np.arange(len(p))
    mean = np.sum(p * x,dtype=float)
    var = np.sum((x - mean)**2 * p)
    return mean, var

mean, var = mean_var(posterior)
#book_plots.bar_plot(posterior)
plt.bar(xvals, posterior)
plt.plot(xs, gaussian_pdf(xs, mean, var), c='r');
print('mean: %.2f' % mean, 'var: %.2f' % var)
```

# t-student distribution
* this is a variant of normal distribution 
* to fully underestand it, you need to know the degree of freedem
* the more df, the closer to normal (Gaussian distribution)
* All the graphs below are symetricall to the mean line but the tail of each distribution is different.  
* The tails can be shortened, fatter, thinner, or otherwise shaped differently from an exponential distribution. The measure of this is called [*kurtosis*](https://en.wikipedia.org/wiki/Kurtosis). 
* keep in mind that Kurtosis is a the forth order statictis, suitable for describing symmetrical distriubutions

The kurtosis of a distribution is calculated using the following formula:
$$
\begin{equation}
\text{Kurtosis} = \frac{1}{n} \sum_{i=1}^{n} \left(\frac{x_i - \mu}{\sigma}\right)^4 = \mathbb{E}[(\frac{X- \mu}{\sigma})^4] 
\end{equation}
$$
- Other formulas also exist in scientific literature 
- $x_i$ : represents each individual data point,
- $\mu$ : is the mean of the distribution,
- $\sigma$ : is the standard deviation of the distribution,
- $n$ : is the number of data points.
- $\mathbb{E}$ : is the Expected value.



```python
#An example of t-student distribution
# Define the degrees of freedom for the t-distributions
dfs = [1, 2, 3, 10, 30]

# Generate x values for plotting
x = np.linspace(-8, 8, 1000)

# Initialize a plot
plt.figure(figsize=(12, 8))

# Plot the standard normal distribution for comparison
normal_pdf = (1/np.sqrt(2 * np.pi)) * np.exp(-0.5 * x**2)
plt.plot(x, normal_pdf, 'r-', lw=2, label='Normal Distribution')

# Generate and plot the PDFs for different degrees of freedom
for df in dfs:
    # Generate t-distributed data using the PDF formula
    pdf = (np.exp(np.log(1 + (x**2) / df) * -(df + 1) / 2) / 
           (np.sqrt(df) * np.sqrt(np.pi) * np.exp(np.log(np.math.gamma(df/2)) - np.log(np.math.gamma((df + 1) / 2)))))
    
    plt.plot(x, pdf, lw=2, label=f't-Distribution (df={df})')
    # Calculate kurtosis for the t-distribution
    kurtosis = 6 / (df - 4) if df > 4 else np.inf  # Excess kurtosis
    
    # Annotate the plot with kurtosis value
    if kurtosis != np.inf:
        plt.text(3, 0.35 - 0.05 * dfs.index(df), f'Kurtosis (df={df}): {kurtosis:.2f}', fontsize=12, color='black')
    else:
        plt.text(3, 0.35 - 0.05 * dfs.index(df), f'Kurtosis (df={df}): Undefined', fontsize=12, color='black')

plt.axvline(0, color='k', linestyle='--', linewidth=1, label ='mean value')

# Add titles and labels
plt.title('Comparison of Student\'s t-Distributions with Different Degrees of Freedom')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

# Log-normal distribution
* this is a variant of normal distribution 
* to fully underestand it, you should pay attention that it is symmetrical around mean (or Zero) since it cannot take negative numbers, as thw word of 'log' implies
* the measure of asymmetrical property is [*skew*](https://en.wikipedia.org/wiki/Skewness)

$$
\begin{equation}
\text{Skewness} = \frac{1}{n} \sum_{i=1}^{n} \left(\frac{x_i - \mu}{\sigma}\right)^3 = \mathbb{E}[(\frac{X- \mu}{\sigma})^3] 
\end{equation}
$$

```python
def lognormal_pdf(x, mean, sigma):
    """
    Compute the probability density function (PDF) of a log-normal distribution.

    Parameters:
    x (float or np.ndarray): The point(s) at which to evaluate the PDF.
    mean (float): The mean (μ) of the underlying normal distribution.
    sigma (float): The standard deviation (σ) of the underlying normal distribution.

    Returns:
    float or np.ndarray: The value of the PDF at x.
    """
    pdf = (np.exp(-(np.log(x) - mean)**2 / (2 * sigma**2))
       / (x * sigma * np.sqrt(2 * np.pi)))
    return pdf

def calculate_skewness(data):
    """
    Calculate the skewness of a dataset.

    Parameters:
    data (np.ndarray): A 1D array of numerical data.

    Returns:
    float: The skewness of the data.
    """
    mean = np.mean(data)
    std_dev = np.std(data)
    n = len(data)
    skewness = np.sum((data - mean)**3) / n / (std_dev**3)
    return skewness

def calculate_mode(data):
    """
    Calculate the mode of a numpy array using numpy functions.

    Parameters:
    data (np.ndarray): A 1D array of numerical data.

    Returns:
    mode (float): The mode of the data.
    count (int): The count of the mode.
    """
    # Get unique values and their counts
    values, counts = np.unique(data, return_counts=True)
    
    # Find the index of the maximum count
    max_count_index = np.argmax(counts)
    
    # Get the mode and its count
    mode_value = values[max_count_index]
    mode_count = counts[max_count_index]
    
    return mode_value
# Define the parameters for the log-normal distribution
mean = [1, 1.5, 2]
sigma = [0.1, 0.3, 0.5]

# Generate x values for plotting the PDF
x_values = np.arange(0,100,0.01) # Avoid x=0 since log-normal is undefined at x=0

# Compute the PDF using the defined function
for i in range(3):
    pdf_values = lognormal_pdf(x_values, mean[i], sigma[i])

# Plot the PDF
    plt.plot(x_values, pdf_values, lw=2, label=r'log-normal pdf ($\sigma =$ {})'.format(sigma[i]))
    skew = calculate_skewness(pdf_values)
    print(f'the skewness values is {skew}')

#plt.axvline(1.25, color='blue', linestyle='--', linewidth=2, label=f'Mean: {mean[2]:.2f}')
#median = np.median(pdf_values)
#mode = calculate_mode(pdf_values)
#plt.plot(median, color='g', linestyle='--', linewidth=2, label=f'Median: {median:.2f}')
#plt.plot(mode, color='g', linestyle='--', linewidth=2, label=f'Mode: {mode:.2f}')


plt.title('Log-normal Distribution PDF')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.xlim([0,10])
plt.show()
```
